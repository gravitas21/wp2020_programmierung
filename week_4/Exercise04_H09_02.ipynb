## Exercise 04. Homework 09. Point 02

For $i = 1, 2, \ldots, 10$, generate $N = 10^{5}$ random numbers for each $x^{\left(i \right)}_{n}$, which are equally distributed between $0$ and $1$ ($n = 1,2, \ldots, N$). From this, generate the sums $u^{\left(z\right)}_{n} = \sum^{z}_{i = 1} x^{\left(i\right)_{n}}$ for $z = 2; 3; 10$. Determine numerically the probability densities of $x^{\left(1\right)}$ and $u^{\left(z\right)}$ and plot them together with the appropriate Gaussian distribution. What does the result illustrate?

```python
def gaussian_distribution(mean, variance, x):
    return (2.* np.pi * variance) ** (-.5) * np.exp(-(x-mean) ** 2 /(2 * variance))
```

```python
# Initial values

N = 10 ** 5
mean_1 = 0.5
variance_1 = 1/12
```

```python
# Two uniform distributions
# Figure
fig = plt.figure(figsize=(16, 9))

N_uniform = 2
rand_x = np.zeros(N)
    
for _ in range(N_uniform):
        
    rand_x += np.random.random(size=N)
        
hist, bins = np.histogram(rand_x, bins=np.linspace(rand_x.min() - 0.4, rand_x.max() + 0.4, 30),
                          density=True)
plt.plot((bins[1:]+bins[:-1])/2, hist, linewidth=5, label=r'$x_i={}$'.format(N_uniform))

var = variance_1 * N_uniform
mean = mean_1 * N_uniform

plt.plot(bins, gaussian_distribution(mean, var, bins), '--k', label='Gaussian distribution')

plt.xlabel(r'$u^{(%d)}$' % N_uniform, fontsize=20)
plt.ylabel('pdf', fontsize=20)
plt.legend(loc='best', fontsize=20)
plt.grid(True)
```
