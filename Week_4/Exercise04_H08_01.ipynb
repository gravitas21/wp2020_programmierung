## Exercise 04. Homework 08. Point 01

The choice $A\left(S,t\right) = B\left(S,t\right) = 1$ for the coefficient functions defines the Brownian motion with drift

$$dS=\mu\,dt+\sigma\varepsilon\,\sqrt{dt}$$

Choose an appropriate time discretization of $t_{0}$ to $t_{1}$ in $N$ steps and generate different runs of the simulation with unchanged parameters.

```python
def brownian_motion(t0, t, s0, mu, sigma, N):
    
    delta_t = (t - t0) / N
    
    epsilon = np.random.normal(size=N) 
    
    delta_s = mu * delta_t * np.ones(N) + sigma * epsilon * delta_t ** .5
    
    t_vec = np.append(t0, t0 + np.cumsum(delta_t * np.ones(N)))
    s_vec = np.append(s0, s0 + np.cumsum(delta_s))
    
    return (t_vec , s_vec)
```

```python
# Initial values

t0 = 0
s0 = 1
mu = 1
sigma = 1
t = 1
N = 1000

simulations = 6
```

```python
# Brownian motion

# Figure
fig = plt.figure(figsize=(16,9))

for sim in range(simulations):
    
    (t_vec, s_vec) = brownian_motion(t0, t, s0, mu, sigma, N)
    plt.plot(t_vec, s_vec, label=f'Simulation {sim}')
    
# Clean drift (sigma = 0)
(t_vec, s_vec) = brownian_motion(t0, t, s0, mu, 0 * sigma, N)
plt.plot(t_vec, s_vec, '--k', linewidth = 3, label=r'Clean drift ($\sigma = {}$)'.format(0 * sigma))

plt.xlabel("$t$", fontsize=20)
plt.ylabel('$S$', fontsize=20)
plt.title(r"$\mu={}$, $\sigma={}$, $N={}$, $t_0 = {}$, $s(t_0)= {}$".format(mu, sigma, N, t0, s0), fontsize=20)
plt.legend(loc='best')
plt.xlim(min(t_vec), max(t_vec))
plt.grid(True)
```

```python
# Different number of increments
N_vec = [10, 100, 1000, 10000]

# Figure
fig = plt.figure(figsize=(16,9))

for N in N_vec:
    
    (t_vec, s_vec) = brownian_motion(t0, t, s0, mu, sigma, N)
    plt.plot(t_vec, s_vec, label=f'Increment {N}')

# clean drift (sigma = 0)
(t_vec, s_vec) = brownian_motion(t0, t, s0, mu, 0 * sigma, N)
plt.plot(t_vec, s_vec, '--k', linewidth = 3, label=r'Clean drift ($\sigma = {}$)'.format(0 * sigma))

plt.xlabel("$t$", fontsize=20)
plt.ylabel('$S$', fontsize=20)
plt.title(r"$\mu={}$, $\sigma={}$, $t_0 = {}$, $s(t_0)= {}$".format(mu, sigma, t0, s0), fontsize=20)
plt.legend(loc='best')
plt.xlim(min(t_vec), max(t_vec))
plt.grid(True)
```
