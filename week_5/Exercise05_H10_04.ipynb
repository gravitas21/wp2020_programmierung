## Exercise 05. Homework 10. Point 04

Furthermore, the betting office knows the probability $p$ for the output $A$. Then it can always set $q$ such that either $p\le q \le v$ or $v \le q \le p$. Calculate the expected profit of the betting office divided by the total wagers. Draw this as a function of $q$ for $p = 0. 2$ and $v = 0.4$ in the range $p \le q \le v$.

```python
def rel_gain(p,v,q):
    return 1. - p * v / q - (1. - p) * (1.- v) / (1. - q)
```

```python
# Initial values

p = 0.2
v = 0.4
q = np.linspace(p,v, 1000)
```

```python
# Figure
fig = plt.figure(figsize=(16,9))

plt.plot(q, rel_gain(p,v,q), linewidth=10, label=r'Relative gain')
plt.legend(fontsize=30)
plt.xlabel(r'$q$', fontsize=40)
plt.ylabel(r'$\frac{G}{E}$', fontsize=40)
plt.grid(True)
plt.tight_layout()
```
