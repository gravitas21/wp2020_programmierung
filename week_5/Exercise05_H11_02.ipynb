## Exercise 05. Homework 11. Point 02

Generate $N = 10^3$ Cauchy distributed random numbers and Calculate $\max_{n} \left| u_{n} \right|$, standard deviation, mean and median of the sample. Repeat this for $N = 10^4$ and $N = 10^5$. How is the mean distributed according to the lecture?

```python
N_val = [1000, 10000, 100000]

print('Cauchy distributed random numbers')
print()

for N in N_val:
    
    x = np.random.random(size=N)
    u = 1 / np.tan(np.pi * x)
    u_n = np.abs(u).max()
    
    
    print(f'N = {N}')
    print(f'u_n(max) = {u_n:.2f}, std = {np.std(u):.2f}, mean = {np.mean(u):.2f}, median = {np.median(u):.4f}')
    print()
```

```python
# Initial values
Ns = np.array(np.logspace(2,7,10000))
maxs = 0. * Ns
means = 0. * Ns
meds = 0. * Ns 
stds = 0. * Ns
```

```python
# Values

for i, N_ in enumerate(Ns):

    N = int(N_)
    x = np.random.random(size=N)
    u = 1. / np.tan(np.pi * x)
    maxs[i] = np.abs(u).max()
    means[i] = np.mean(u)
    meds[i] = np.median(u)
    stds[i] = np.std(u)
```

```python
# Figure
fig = plt.figure(figsize=(16,9))

plt.loglog(Ns, maxs, '-+', label=r'$max_{n}\left|u_{n}\right| \sim N$')
plt.loglog(Ns, stds, '-+', label=r'$std \sim \sqrt{N}$')
plt.loglog(Ns, np.abs(means), '-+', label=r'$|mean| \sim N^{0} = Constant$')
plt.loglog(Ns, np.abs(meds), '-+', label=r'$|median| \sim \frac{1}{\sqrt{N}}$')

plt.legend(fontsize=20)
plt.xlabel(r'$N$', fontsize=40)
plt.grid(True)
plt.tight_layout()
```
