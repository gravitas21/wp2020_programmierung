## Exercise 05. Homework 11. Point 03

Create and draw the time series of the L\'{e}vy-flight
    
$$ U \left( t \right) = \sum_{n=1}^{\frac{t}{\Delta t}} u_{n} $$
    
to $\frac{t_{max}}{\Delta t} = 10^{5}$. How does $U \left(t\right)$ differ from an ordinary random walk?

```python
# Initial values

N = 10 ** 5
simulations = 3
```

```python
# Figure
fig = plt.figure(figsize=(16,9))

for sim in range(simulations):
    
    # Random numbers
    x = np.random.random(size=N)
    # Levy-flight
    u = 1. / np.tan(np.pi * x)
    
    plt.plot(np.cumsum(u), label='Sim {}'.format(sim))

plt.legend(loc='best', fontsize=20)
plt.xlabel(r'$t$', fontsize=20)
plt.ylabel(r'$U$', fontsize=20)
plt.title('$\Delta t=1$', fontsize=20)
plt.grid(True)
plt.tight_layout()
```
