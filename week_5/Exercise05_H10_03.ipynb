## Exercise 05. Homework 10. Point 03

Assuming a betting office can predict the total bets $E_{A} \left(Q_{A}\right)$ and $E_{\overline{A}}\left(Q_{\overline{A}}\right)$ depending on the odds. You can assume that the relative volume $v = \frac{E_{A}}{E_{A} + E_{\overline{A}}}$ as a function of the inverse quota $q = \frac{1}{Q_{A}}$ is a smooth monotone decreasing function with $v \left(q \rightarrow 0\right) = 1$ and $v \left(q \rightarrow 1\right) = 0$. Sketch $v \left(q\right)$ and graphically determine $q_{0}$ with $q_{0} = v \left(q_{0}\right)$. Show that with $Q_{A} = \frac{1}{q_{0}}$ the betting office does not take any risk, i.e. it pays the same amount regardless of the outcome.

```python
def gaussian_distribution(mean, variance, x):
    return (1 / (2 * np.pi * variance) ** 0.5) * np.exp(-((x - mean) ** 2) / (2 * variance))
```

```python
# Initial values

mean = 0
variance = 0.16

x = np.arange(0, 1, 0.0001)
gaussian = gaussian_distribution(mean, variance, x)
```

```python
# Sketch
# Remember that this is just a sketch, the real shape of the
# theoretical function is not used.
fig = plt.figure(figsize=(16,9))

# Smooth monotone decreasing function
plt.plot(x, gaussian, linewidth=5, label=r'$v(q)$')
# q_0 = v(q_0)
plt.plot(x,x, linewidth=5, label=r'$v(q) = q$')

# Find the intersection
idx = np.argwhere(np.diff(np.sign(gaussian - x))).flatten()
# Dashed line to the x and y axis
plt.plot([x[idx], x[idx]], [0, gaussian[idx]], '--k')
plt.plot([x[idx], 0], [gaussian[idx], gaussian[idx]], '--k')
# Plot the intersection
plt.plot(x[idx], gaussian[idx], 'ro', markersize=20, label=r'$v(q_0) = q_0$')

# Figure details
plt.legend(loc='best', fontsize=30)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel(r'$q$', fontsize=25)
plt.ylabel(r'$v$', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
```
