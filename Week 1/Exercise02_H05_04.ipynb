## Exercise 02. Homework 05. Point 04

Were you going to participate in the game if you had to use a fifth of the balance $G \left(t-1\right)$ in each roundtable? Simulate the gameplay five times.

```python
# Initial values

rounds = 500
percentage = 1/5
bet_0 = 100.
simulations = 5
win_factor = 1.5
lose_factor = 0.6
```

```python
fig1 = plt.figure(figsize=(16,9))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

for sim in range(simulations):
    
    gains = bet_0 * np.ones(rounds)
    gains[0] = bet_0
    
    for round_ in range(1,rounds):
        
        # get an integer random value (0 or 1)
        rand_val = np.random.randint(2)
        # Conserve 4/5 off the previous round and bet only
        # 1/5 of the previous round
        gains[round_] = gains[round_ - 1] * (1 - percentage) \
            + percentage * gains[round_ - 1] \
            * ((rand_val == 0) * lose_factor + (rand_val == 1) * win_factor)
        
    # Plot
    ax1.plot(gains, label=f'Simulation {sim + 1}')
    ax2.semilogy(gains, label=f'Simulation {sim + 1}')
    
# Figure details
ax1.set_xlim(0,500)
ax1.legend(loc='best', fontsize=15)
ax1.set_title('', fontsize=20)
ax1.set_xlabel(r'Rounds', fontsize=20)
ax1.set_ylabel(r'Gains $[euro]$', fontsize=20)
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)
ax1.grid(True)

ax2.axhline(0.01, color='black', linewidth=5)
ax2.set_xlim(0,500)
ax2.legend(loc='best', fontsize=15)
ax2.set_title('', fontsize=20)
ax2.set_xlabel(r'Rounds', fontsize=20)
ax2.set_ylabel(r'Gains $[euro]$', fontsize=20)
ax2.tick_params(axis='x', labelsize=15)
ax2.tick_params(axis='y', labelsize=15)
ax2.grid(True)
```
