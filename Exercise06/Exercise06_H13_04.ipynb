## Exercise 06. Homework 13. Point 04

For $K = 4$ companies, generate the artificial time series $G_{k} \left(t\right) = x_{k, t}$ the length $T = 10$, with independent, standard normal distributed random numbers $x_{k, t}$. Calculate the correlation coefficients $C_{kl}$. Repeat for $T = 100$ and $T = 1000$. Discuss the results briefly.

```python
# Initial values

K=20
T = [10,100,1000,10000]
```

```python
# Figure
fig = plt.figure(figsize=(16,9))
plt.subplots_adjust(hspace=0.3, wspace=0)



for T_idx,T_val in enumerate(T):
    G = np.random.normal(size=(K,T_val))
    G_mean = np.mean(G, axis=1)
    G_std = np.std(G, axis=1)
    C = (np.dot(G,G.transpose()) / len(G[0])
         - np.tensordot(G_mean,G_mean,axes=0)) / np.tensordot(G_std,G_std,axes=0)
    
    plt.subplot(2, 2, T_idx + 1)
    plt.title(r'$T={}$'.format(T_val))
    plt.xlabel('$k$')
    plt.ylabel('$k$')
    plt.imshow(C, vmin=-1, vmax=1, cmap=plt.get_cmap("bwr_r"))
    plt.colorbar()
```
