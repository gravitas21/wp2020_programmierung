```python
import numpy as np
from matplotlib import pyplot as plt

from ipywidgets import interactive
```

## Exercise 02. Homework 04. Point 01
Sketch $p^{\left( C \right)}\left( x \right)$ for different values of $\gamma$.

The Lorentz or Cauchy distribution is given by

$$p^{\left(C\right)}\left(x\right)=\frac{\gamma}{\pi}\cdot\frac{1}{x^{2}+\gamma^{2}}$$

where $\gamma$ is the width.

```python
def lorentz_cauchy_distribution(gamma, x):
    return (gamma / np.pi) * (1 / (x * x + gamma * gamma))
```

```python
# Initial values

# Linear plot
x1 = np.arange(-5, 5, 0.01)
# Log plot
x2 = np.arange(0, 1000, 0.01)
gammas = [0.5, 1, 2, 10]
```

```python
fig = plt.figure(figsize=(16, 9))
plt.subplots_adjust(hspace=0, wspace=0)
  
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

for gamma in gammas:
    
    # Plot
    ax1.plot(x1, lorentz_cauchy_distribution(gamma, x1), label='$\gamma$='+str(gamma))
    ax2.loglog(x2, lorentz_cauchy_distribution(gamma, x2), label='$\gamma$='+str(gamma))

# Figure details
ax1.set_xlabel(r'x', fontsize=20)
ax1.set_ylabel(r'$p^{\left( C \right)}\left( x \right)$', fontsize=20)
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)
ax1.set_xlim(-5,5)
ax1.legend(loc='best', fontsize=15)
ax1.set_title('Lorentz or Cauchy distribution', fontsize=20)
ax1.grid(True)
plt.tight_layout()

ax2.set_xlabel(r'x', fontsize=20)
ax2.set_ylabel(r'$p^{\left( C \right)}\left( x \right)$', fontsize=20)
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)
ax2.set_xlim(10**-1,10**3)
ax2.legend(loc='best', fontsize=15)
ax2.set_title('Lorentz or Cauchy distribution - Log Log plot', fontsize=20)
ax2.grid(True)
plt.tight_layout()
```

```python
# Initial values
gammas = [0.0001, 0.001, 0.01, 0.5, 1, 2, 10, 20]
```

```python
def lorentz_cauchy_distribution_plot(gamma):
    
    fig = plt.figure(figsize=(16, 9))

    x = np.arange(-10, 10, 0.01)
    p = lorentz_cauchy_distribution(gamma, x)
    
    # Plot
    plt.plot(x, p,label='$\gamma$='+str(gamma))
    plt.xlabel(r'x', fontsize=20)
    plt.ylabel(r'$p^{\left( C \right)}\left( x \right)$', fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.xlim(-10,10)
    plt.legend(loc='best', fontsize=15)
    plt.title('Lorentz or Cauchy distribution', fontsize=20)
    plt.grid(True)
    plt.tight_layout()
```

```python
interactive(lorentz_cauchy_distribution_plot, gamma=gammas)
```
