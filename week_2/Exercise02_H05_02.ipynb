## Exercise 02. Homework 05. Point 02

The time-dependent credit $G \left(t \right)$ is initially at $G \left(0\right) = 100 \, Euro$. You always have to use all your credit. Thus the credit $G \left(1\right)$ after the first round, depending on the toss, is either $60 \, Euro$ or $150 \, Euro$. In the second round you must complete $G \left(1\right)$, and so on. There are a total of $500$ laps, all of which you must complete. Would you like to participate in the game? Simulate as a decision aid five times the gameplay with all $500$ rounds. Use a suitable random number generator. Display the credit as a function of time for all five passes in the same figure, with the y-axis shown logarithmically.

```python
# Initial values

rounds = 500
bet_0 = 100.
simulations = 5
win_factor = 1.5
lose_factor = 0.6
```

```python
fig1 = plt.figure(figsize=(16,9))
ax1 = plt.subplot(121)
ax2 = plt.axes([.236, .4, .23, .23])
ax3 = plt.subplot(122)

for sim in range(simulations):
    
    gains = np.zeros(rounds)
    gains[0] = bet_0
    
    for round_ in range(1, rounds):
        
        # Get an integer random value (0 or 1)
        rand_val = np.random.randint(2)
        gains[round_] = gains[round_ - 1] \
            * ((rand_val == 0) * lose_factor + (rand_val == 1) * win_factor)
        
    # Plot
    ax1.plot(gains, label=f'Simulation {sim + 1}')
    ax2.plot(gains[:100])
    ax3.semilogy(gains, label=f'Simulation {sim + 1}')
    
# Figure details
ax1.set_xlim(0,500)
ax1.legend(loc='best', fontsize=15)
ax1.set_xlabel(r'Rounds', fontsize=20)
ax1.set_ylabel(r'Gains $[euro]$', fontsize=20)
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)
ax1.grid(True)

plt.axhline(0.01, color='black', linewidth=5)
ax2.set_title('Zoom', fontsize=15)
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 100)
ax2.set_xlabel(r'Rounds', fontsize=10)
ax2.set_ylabel(r'Gains $[euro]$', fontsize=10)
ax2.grid(True)

ax3.set_xlim(0,500)
ax3.legend(loc='best', fontsize=15)
ax3.set_xlabel(r'Rounds', fontsize=20)
ax3.set_ylabel(r'Gains $[euro]$', fontsize=20)
ax3.tick_params(axis='x', labelsize=15)
ax3.tick_params(axis='y', labelsize=15)
ax3.grid(True)
```
